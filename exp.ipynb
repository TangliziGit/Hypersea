```python
import matplotlib.pyplot as plt
import pickle as pkl
```

```python
losses = pkl.load(open('pkl/losses.pkl', 'rb'))

k = 20
loss = losses[0]
ls = [sum(loss[i:i+k])/len(loss[i:i+k]) for i in range(0, len(loss), k)]

plt.figure(figsize=(12, 6))
plt.plot(ls)
```
